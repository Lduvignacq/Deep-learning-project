### 🎯 C'est quoi DEX?

**DEX** (Deep EXpectation) est une révolution en estimation d'âge par vision par ordinateur qui reformule le problème :

- **Avant (Régression classique)**: Prédire l'âge = 1 nombre
- **Après (DEX - Classification)**: Prédire la probabilité pour chaque âge, puis calculer la moyenne pondérée

### 💡 Équation centrale de DEX:

$$\hat{\text{age}} = \sum_{i=0}^{100} i \times P(\text{age} = i)$$

Cette formule simple mais puissante transforme le problème en **classification multi-classe** (101 classes).

### 🏆 Résultats impressionnants:

- **MAE**: 4-5 années (vs humans 6-8 années) → **Performance surhumaine**
- **Dataset**: IMDB-WIKI (500k+ images) → **Power of big data**
- **Architecture**: VGG-16 + fine-tuning → **Transfer learning**

````python
"""
TEMPLATE COMPLET: Entraînement, Évaluation et Inférence DEX
"""

class DEXPipeline:
    """Pipeline complet pour DEX: Train → Eval → Predict"""
    
    def __init__(self, model_path=None, device='cuda'):
        self.device = torch.device(device if torch.cuda.is_available() else 'cpu')
        self.model = DEXModel(num_classes=AGE_BINS, pretrained=True).to(self.device)
        self.model_path = model_path
        
        if model_path and os.path.exists(model_path):
            self.load_checkpoint(model_path)
    
    def train(self, train_images, train_ages, val_images=None, val_ages=None,
              epochs=10, batch_size=32, **kwargs):
        """Entraîne le modèle DEX"""
        
        # Créer les dataloaders
        train_dataset = AgeDataset(train_images, train_ages, transform=train_transform)
        train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
        
        if val_images is not None:
            val_dataset = AgeDataset(val_images, val_ages, transform=test_transform)
            val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
        else:
            val_loader = None
        
        # Optimizer
        optimizer = optim.Adam(self.model.parameters(), lr=0.001)
        
        history = {'train_loss': [], 'train_mae': [], 'val_loss': [], 'val_mae': []}
        
        for epoch in range(epochs):
            # Training
            train_loss, train_mae = train_epoch(self.model, train_loader, optimizer, self.device)
            history['train_loss'].append(train_loss)
            history['train_mae'].append(train_mae)
            
            # Validation
            if val_loader:
                val_loss, val_mae, _, _ = evaluate(self.model, val_loader, self.device)
                history['val_loss'].append(val_loss)
                history['val_mae'].append(val_mae)
                
                print(f"Epoch {epoch+1}/{epochs} | "
                      f"Train MAE: {train_mae:.2f} | Val MAE: {val_mae:.2f}")
            else:
                print(f"Epoch {epoch+1}/{epochs} | Train MAE: {train_mae:.2f}")
        
        return history
    
    def predict_from_image(self, image_path):
        """Prédit l'âge d'une image"""
        result = predict_age(self.model, image_path, self.device)
        return result
    
    def predict_batch(self, image_paths):
        """Prédit les âges d'un batch d'images"""
        return predict_age_batch(self.model, image_paths, self.device)
    
    def save_checkpoint(self, path):
        """Sauvegarde le modèle"""
        torch.save(self.model.state_dict(), path)
        print(f"✅ Modèle sauvegardé : {path}")
    
    def load_checkpoint(self, path):
        """Charge un modèle entraîné"""
        self.model.load_state_dict(torch.load(path, map_location=self.device))
        self.model.eval()
        print(f"✅ Modèle chargé : {path}")


# Exemple d'utilisation
print("=" * 70)
print("TEMPLATE: DEX PIPELINE COMPLET")
print("=" * 70)

example_code = '''
# 1. Initialiser le pipeline
dex = DEXPipeline(device='cuda')

# 2. Charger les données APPA-REAL
train_images, train_ages = load_appa_real_dataset("appa-real-release", "train")
val_images, val_ages = load_appa_real_dataset("appa-real-release", "valid")
test_images, test_ages = load_appa_real_dataset("appa-real-release", "test")

# 3. Entraîner le modèle
history = dex.train(train_images, train_ages, val_images, val_ages, epochs=30, batch_size=64)

# 4. Sauvegarder
dex.save_checkpoint("dex_model.pth")

# 5. Évaluer sur test set
test_loss, test_mae, predictions, gt = evaluate(dex.model, test_loader, device)
print(f"Test MAE: {test_mae:.2f} years")

# 6. Prédire sur une nouvelle image
result = dex.predict_from_image("path/to/image.jpg")
print(f"Estimated age: {result['estimated_age']:.1f} years")
print(f"Confidence: {result['confidence']:.2%}")

# 7. Analyser la performance
fig, metrics = plot_performance(predictions, test_ages)
'''

print("\nExemple de code complet:")
print(example_code)
print("\n✅ Pipeline DEX prêt à être utilisé !")
````

## 10. Points clés de la méthode DEX

### ✅ Avantages de DEX vs Régression directe:

| Aspect | Régression | DEX |
|--------|-----------|-----|
| **Approche** | Prédire directement l'âge (1 valeur) | Prédire distribution d'âges (101 values) |
| **Loss Function** | MSE/L2 Loss | Cross-Entropy (classification) |
| **Stabilité** | Moins stable | Plus stable, moins sensible aux outliers |
| **Interprétabilité** | Valeur brute | Distribution complète + expected value |
| **Performance** | ≈ human level | **Surhumaine** |

### 🎯 Architecture DEX:

```
Input Image (227×227)
    ↓
[CNN Feature Extraction] (VGG-16 pretrained)
    ↓
[Global Average Pooling]
    ↓
[Fully Connected Layers with Dropout]
    ├─ FC: 4096 → ReLU → Dropout
    ├─ FC: 4096 → ReLU → Dropout
    └─ FC: 4096 → 101 (softmax)
    ↓
[Probability Distribution P(age=0..100)]
    ↓
[Expected Value Calculation]
    └─ E[age] = Σ(i × P(age=i))
    ↓
Predicted Age (float value)
```

### 📊 Métriques clés:

- **MAE (Mean Absolute Error)**: Erreur moyenne en années (target: < 5 ans)
- **CS (Cumulative Score)**: % de prédictions à ±k années
- **RMSE**: Penalise plus les gros erreurs

### 🚀 Améliorations possibles:

1. **Augmentation de données avancée**: Utilisez albumentations pour plus de robustesse
2. **Fine-tuning**: Adaptez les poids VGG-16 sur votre dataset spécifique
3. **Ensemble methods**: Combinez plusieurs modèles pour meilleure performance
4. **Regularization**: L1/L2, Dropout, BatchNorm pour éviter l'overfitting
5. **Hard example mining**: Focus sur des images difficiles

### 📚 Références:

- **Papier original**: Rothe et al., "DEX: Deep EXpectation of Apparent Age from a Single Image" (ICCV 2015)
- **Dataset IMDB-WIKI**: 500k+ d'images annotées en âge
- **Performance**: MAE ≈ 4-5 ans (vs human ≈ 6-8 ans)


````python
def compute_metrics(predictions, ground_truths):
    """
    Calcule les métriques de performance pour l'estimation d'âge.
    
    Métriques:
    - MAE: Mean Absolute Error (erreur moyenne en années)
    - RMSE: Root Mean Square Error
    - CS (Cumulative Score): % de prédictions à ±k ans
    """
    mae = np.mean(np.abs(predictions - ground_truths))
    rmse = np.sqrt(np.mean((predictions - ground_truths) ** 2))
    
    # Cumulative Score pour différents seuils
    cs_5 = np.mean(np.abs(predictions - ground_truths) <= 5)
    cs_10 = np.mean(np.abs(predictions - ground_truths) <= 10)
    
    return {
        'MAE': mae,
        'RMSE': rmse,
        'CS_5': cs_5,      # % correct à ±5 ans
        'CS_10': cs_10,    # % correct à ±10 ans
    }


def plot_performance(predictions, ground_truths, title="DEX Model Performance"):
    """Visualise les performances du modèle"""
    
    metrics = compute_metrics(predictions, ground_truths)
    
    fig = plt.figure(figsize=(16, 10))
    gs = GridSpec(2, 3, figure=fig, hspace=0.3, wspace=0.3)
    
    # 1. Scatter plot: Predicted vs Ground Truth
    ax1 = fig.add_subplot(gs[0, 0])
    ax1.scatter(ground_truths, predictions, alpha=0.5, s=30)
    min_age, max_age = ground_truths.min(), ground_truths.max()
    ax1.plot([min_age, max_age], [min_age, max_age], 'r--', linewidth=2, label='Perfect prediction')
    ax1.set_xlabel('Ground Truth Age')
    ax1.set_ylabel('Predicted Age')
    ax1.set_title('Predicted vs Ground Truth')
    ax1.legend()
    ax1.grid(alpha=0.3)
    
    # 2. Error distribution
    ax2 = fig.add_subplot(gs[0, 1])
    errors = predictions - ground_truths
    ax2.hist(errors, bins=30, edgecolor='black', color='steelblue', alpha=0.7)
    ax2.axvline(0, color='red', linestyle='--', linewidth=2)
    ax2.set_xlabel('Prediction Error (years)')
    ax2.set_ylabel('Frequency')
    ax2.set_title(f'Error Distribution (MAE: {metrics["MAE"]:.2f} years)')
    ax2.grid(alpha=0.3)
    
    # 3. Absolute Error by Age
    ax3 = fig.add_subplot(gs[0, 2])
    abs_errors = np.abs(errors)
    ax3.scatter(ground_truths, abs_errors, alpha=0.5, s=30)
    ax3.axhline(metrics['MAE'], color='red', linestyle='--', linewidth=2, label=f'Mean MAE: {metrics["MAE"]:.2f}')
    ax3.set_xlabel('Ground Truth Age')
    ax3.set_ylabel('Absolute Error (years)')
    ax3.set_title('Absolute Error by Age')
    ax3.legend()
    ax3.grid(alpha=0.3)
    
    # 4. Cumulative Score
    ax4 = fig.add_subplot(gs[1, 0])
    thresholds = range(0, 21)
    cs_scores = [np.mean(np.abs(predictions - ground_truths) <= t) for t in thresholds]
    ax4.plot(thresholds, cs_scores, marker='o', linewidth=2, markersize=6)
    ax4.fill_between(thresholds, cs_scores, alpha=0.3)
    ax4.set_xlabel('Threshold (years)')
    ax4.set_ylabel('Cumulative Score')
    ax4.set_title('Cumulative Score (%)')
    ax4.grid(alpha=0.3)
    ax4.set_ylim([0, 1])
    
    # 5. Metrics table
    ax5 = fig.add_subplot(gs[1, 1])
    ax5.axis('off')
    
    metrics_text = f"""
    PERFORMANCE METRICS
    
    MAE (Mean Absolute Error): {metrics['MAE']:.2f} years
    RMSE (Root Mean Squared): {metrics['RMSE']:.2f} years
    
    Cumulative Score:
    ├─ ±5 years:  {metrics['CS_5']*100:.1f}%
    └─ ±10 years: {metrics['CS_10']*100:.1f}%
    
    Samples: {len(predictions)}
    Age range: {ground_truths.min():.0f} - {ground_truths.max():.0f} years
    """
    
    ax5.text(0.1, 0.5, metrics_text, fontsize=11, family='monospace',
             verticalalignment='center', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    # 6. Age distribution comparison
    ax6 = fig.add_subplot(gs[1, 2])
    ax6.hist([ground_truths, predictions], bins=20, label=['Ground Truth', 'Predicted'], alpha=0.7)
    ax6.set_xlabel('Age')
    ax6.set_ylabel('Frequency')
    ax6.set_title('Age Distribution Comparison')
    ax6.legend()
    ax6.grid(alpha=0.3)
    
    fig.suptitle(title, fontsize=14, fontweight='bold')
    
    return fig, metrics


# Test avec des données synthétiques
print("✅ Fonctions d'analyse créées")
````

````python
def load_appa_real_dataset(data_dir, split='train'):
    """
    Charge le dataset APPA-REAL.
    
    Args:
        data_dir: Chemin vers le dossier appa-real-release/
        split: 'train', 'valid', ou 'test'
    
    Returns:
        image_paths, ages (arrays)
    """
    import scipy.io as sio
    
    split_dir = os.path.join(data_dir, split)
    gt_file = os.path.join(data_dir, f'gt_{split}.csv')
    
    # Lire les labels
    df = pd.read_csv(gt_file)
    
    image_paths = []
    ages = []
    
    for idx, row in df.iterrows():
        img_id = row.iloc[0]
        age = row.iloc[1]  # Âge moyen
        
        # Les images dans APPA-REAL sont au format .jpg.mat
        img_path = os.path.join(split_dir, f'{img_id}.jpg')
        
        if os.path.exists(img_path):
            image_paths.append(img_path)
            ages.append(age)
    
    return np.array(image_paths), np.array(ages)


# Exemple d'utilisation
print("✅ Fonction load_appa_real_dataset() créée")
print("""
Utilisation:
    train_images, train_ages = load_appa_real_dataset('appa-real-release', split='train')
    valid_images, valid_ages = load_appa_real_dataset('appa-real-release', split='valid')
    test_images, test_ages = load_appa_real_dataset('appa-real-release', split='test')
""")
````

````python
def predict_age(model, image_path, device):
    """
    Prédit l'âge à partir d'une image.
    
    Retourne:
    - estimated_age: L'âge estimé (weighted average)
    - age_distribution: La distribution de probabilité pour chaque âge
    - confidence: La confiance (probabilité max)
    """
    model.eval()
    
    # Charger et prétraiter l'image
    image = Image.open(image_path).convert('RGB')
    image_tensor = test_transform(image).unsqueeze(0).to(device)
    
    with torch.no_grad():
        logits, probs = model(image_tensor)
        
        # Âge estimé (weighted average - le cœur de DEX)
        age_range = torch.arange(0, AGE_BINS, dtype=torch.float32, device=device)
        estimated_age = torch.sum(probs * age_range.unsqueeze(0), dim=1).item()
        
        # Distribution de probabilité
        age_dist = probs.cpu().numpy()[0]
        
        # Confiance (probabilité maximale)
        confidence = age_dist.max()
        
        # Âge avec confiance maximale (argmax)
        argmax_age = np.argmax(age_dist)
    
    return {
        'estimated_age': float(estimated_age),
        'argmax_age': int(argmax_age),
        'confidence': float(confidence),
        'age_distribution': age_dist
    }


def predict_age_batch(model, image_paths, device):
    """Prédit les âges pour un batch d'images"""
    results = []
    for img_path in image_paths:
        result = predict_age(model, img_path, device)
        results.append(result)
    return results


def visualize_age_prediction(image_path, prediction_result, title="Age Prediction"):
    """Visualise une prédiction d'âge avec la distribution de probabilité"""
    image = Image.open(image_path).convert('RGB')
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Image
    ax1.imshow(image)
    ax1.set_title(title, fontsize=14, fontweight='bold')
    ax1.axis('off')
    
    # Distribution de probabilité
    age_dist = prediction_result['age_distribution']
    ax2.bar(range(AGE_BINS), age_dist, color='steelblue', alpha=0.7)
    
    # Marquer l'âge estimé
    estimated_age = prediction_result['estimated_age']
    ax2.axvline(estimated_age, color='red', linestyle='--', linewidth=2, label=f'Expected Value: {estimated_age:.1f}')
    
    # Marquer l'âge argmax
    argmax_age = prediction_result['argmax_age']
    ax2.axvline(argmax_age, color='green', linestyle='--', linewidth=2, label=f'ArgMax: {argmax_age}')
    
    ax2.set_xlabel('Age', fontsize=12)
    ax2.set_ylabel('Probability', fontsize=12)
    ax2.set_title('Age Distribution (DEX Method)', fontsize=12, fontweight='bold')
    ax2.legend(fontsize=10)
    ax2.grid(alpha=0.3)
    
    plt.tight_layout()
    return fig

print("✅ Fonctions de prédiction créées")
````

## 6. Pipeline de prédiction (Inférence)

````python
def train_epoch(model, dataloader, optimizer, device):
    """Entraîne le modèle pour une époque"""
    model.train()
    total_loss = 0.0
    total_mae = 0.0
    
    for images, ages in tqdm(dataloader, desc="Training", leave=False):
        images = images.to(device)
        ages = ages.to(device, dtype=torch.float32)
        
        # Forward pass
        logits, probs = model(images)
        
        # Calcul de la loss
        loss = dex_loss(logits, ages)
        
        # Calcul de l'âge estimé
        with torch.no_grad():
            estimated_ages = calculate_expected_age(probs)
            mae = torch.abs(estimated_ages - ages).mean()
        
        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        total_mae += mae.item()
    
    avg_loss = total_loss / len(dataloader)
    avg_mae = total_mae / len(dataloader)
    
    return avg_loss, avg_mae


def evaluate(model, dataloader, device):
    """Évalue le modèle"""
    model.eval()
    total_loss = 0.0
    total_mae = 0.0
    predictions = []
    ground_truths = []
    
    with torch.no_grad():
        for images, ages in tqdm(dataloader, desc="Evaluating", leave=False):
            images = images.to(device)
            ages = ages.to(device, dtype=torch.float32)
            
            # Forward pass
            logits, probs = model(images)
            
            # Calcul de la loss
            loss = dex_loss(logits, ages)
            
            # Calcul de l'âge estimé
            estimated_ages = calculate_expected_age(probs)
            mae = torch.abs(estimated_ages - ages).mean()
            
            total_loss += loss.item()
            total_mae += mae.item()
            
            predictions.extend(estimated_ages.cpu().numpy())
            ground_truths.extend(ages.cpu().numpy())
    
    avg_loss = total_loss / len(dataloader)
    avg_mae = total_mae / len(dataloader)
    
    return avg_loss, avg_mae, np.array(predictions), np.array(ground_truths)

print("✅ Fonctions train_epoch() et evaluate() créées")
````

## 5. Fonctions d'entraînement et d'évaluation

````python
def calculate_expected_age(probs):
    """
    Calcule l'âge attendu (expected value) à partir de la distribution de probabilités.
    
    C'est le cœur de la méthode DEX !
    
    E[age] = Σ(i * P(age=i)) pour i = 0 à 100
    
    Args:
        probs: Tensor de shape (batch_size, 101) avec les probabilités softmax
    
    Returns:
        Tensor de shape (batch_size,) avec les âges estimés
    """
    age_range = torch.arange(0, AGE_BINS, dtype=torch.float32, device=device)
    expected_ages = torch.sum(probs * age_range.unsqueeze(0), dim=1)
    return expected_ages

# Test de la fonction
test_probs = torch.softmax(torch.randn(4, 101), dim=1)
test_ages = calculate_expected_age(test_probs)
print(f"✅ Fonction calculate_expected_age testée")
print(f"   Input shape: {test_probs.shape}")
print(f"   Output shape: {test_ages.shape}")
print(f"   Ages estimés (exemple): {test_ages.cpu().numpy().round(2)}")


# Loss function DEX
criterion_ce = nn.CrossEntropyLoss()

def dex_loss(logits, target_ages):
    """
    Loss function DEX : Cross-entropy sur la distribution des âges.
    
    Les âges sont traités comme 101 classes distinctes (0, 1, 2, ..., 100).
    La loss optimise la prédiction de distribution de probabilités.
    """
    return criterion_ce(logits, target_ages.long())

print("\n✅ Loss function DEX (CrossEntropyLoss) configurée")
````

## 4. Loss Function et Expected Value (Cœur de la méthode DEX)

````python
class DEXModel(nn.Module):
    """
    Modèle DEX (Deep EXpectation) pour l'estimation d'âge.
    
    Architecture:
    - Base: VGG-16 pré-entraîné
    - Output: 101 probabilités (âges 0-100)
    - Loss: Cross-entropy (pour softmax probabilities)
    - Prédiction: Weighted average (expected value)
    """
    
    def __init__(self, num_classes=101, pretrained=True):
        super(DEXModel, self).__init__()
        self.num_classes = num_classes
        
        # Charger VGG-16 pré-entraîné
        vgg16 = models.vgg16(pretrained=pretrained)
        
        # Utiliser les features de VGG-16
        self.features = vgg16.features
        
        # Average pooling
        self.avgpool = vgg16.avgpool
        
        # Remplacer le classificateur pour l'estimation d'âge
        self.classifier = nn.Sequential(
            nn.Linear(25088, 4096),
            nn.ReLU(inplace=True),
            nn.Dropout(p=0.5),
            nn.Linear(4096, 4096),
            nn.ReLU(inplace=True),
            nn.Dropout(p=0.5),
            nn.Linear(4096, num_classes),  # 101 classes (0-100 ans)
        )
        
    def forward(self, x):
        # Features extraction
        x = self.features(x)
        x = self.avgpool(x)
        x = x.view(x.size(0), -1)  # Flatten
        
        # Classification
        logits = self.classifier(x)
        
        # Softmax pour obtenir les probabilités
        probs = torch.nn.functional.softmax(logits, dim=1)
        
        return logits, probs

# Créer le modèle
model = DEXModel(num_classes=AGE_BINS, pretrained=True).to(device)
print(f"✅ Modèle DEX créé avec {AGE_BINS} classes d'âge")
print(f"   Architecture: VGG-16 → 4096 → 4096 → {AGE_BINS}")
print(f"   Loss: CrossEntropyLoss (softmax)")
print(f"   Prédiction: Weighted average (expected value)")
````

## Les 3 piliers de DEX :

### 1. **Deep EXpectation** - Méthode de calcul révolutionnaire
Au lieu de prédire directement une valeur (régression), on prédit une **probabilité pour chaque âge** (0-100 ans) puis on calcule la moyenne pondérée :

$$\hat{age} = \sum_{i=0}^{100} i \cdot P(age=i)$$

Cette approche rend la prédiction beaucoup plus **stable et robuste**.
